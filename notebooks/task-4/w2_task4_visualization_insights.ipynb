```python
# Import Important Librares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

# Load cleaned + analyzed data from task2
df = pd.read_csv("annotated_reviews_task2.csv")
```

```python
#Sentiment Distribution per Bank
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='sentiment_label', hue='bank', palette='Set2')
plt.title("Sentiment Distribution by Bank")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.legend(title='Bank')
plt.show()
```

```python
# Rating Distribution
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='rating', hue='bank', bins=5, multiple='stack', palette='pastel')
plt.title("Rating Distribution by Bank")
plt.xlabel("Star Rating")
plt.ylabel("Count")
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Loop through each unique bank in the dataset
for bank in df['bank'].unique():

    # Filter: Select only the cleaned reviews that are labeled NEGATIVE and are not empty
    text = " ".join(
        df[
            (df['bank'] == bank) &  # Filter for this bank
            (df['sentiment_label'].str.upper() == 'NEGATIVE') &  # Filter for negative sentiment
            (df['cleaned_review'].notna())  # Remove any NaN values in cleaned_review
        ]['cleaned_review']
    )

    #  Check: Proceed only if we have non-empty text for the word cloud
    if text.strip():
        # Create the WordCloud object
        wc = WordCloud(
            width=800,
            height=400,
            background_color='white',
            stopwords=STOPWORDS
        ).generate(text)  # Generate word cloud from text

        #  Display the WordCloud
        plt.figure(figsize=(10, 5))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')  # Remove axis
        plt.title(f"{bank} - Pain Points (NEGATIVE)")  # Add a title
        plt.show()

    else:
        #  Warning: No data to plot
        print(f"⚠️ Skipping {bank}: No negative reviews found.")
```

```python
#Themes by Sentiment (Bar Plot)
plt.figure(figsize=(12,6))
sns.countplot(data=df, y='theme', hue='sentiment_label', order=df['theme'].value_counts().index)
plt.title("Themes by Sentiment")
plt.xlabel("Count")
plt.ylabel("Theme")
plt.legend(title='Sentiment')
plt.show()
```
